# Logistic Regression - Possums

### Import Libraries & Data

```python
# import
import pandas as pd # to import data and work with dataframes
import numpy as np # the backbone to all the data packages
import matplotlib.pyplot as plt # for basic plots
import seaborn as sns # for visualization

# ignore warnings
import warnings
warnings.filterwarnings("ignore")

# to automatically show plots
%matplotlib inline


# import data ----------------

data = pd.read_csv("possum.csv")

data.head()
```

```python
data.info() # like str(data) in R
```

```python
data.describe() # like summary(data) in R
```

### Check for missing data

```python

# Check number of missing values
print("Count of na's:", data.isnull().values.sum())

# Check number of rows with missing values
print("Cases with missing values:", data.isna().any(axis=1).sum())
```

### Drop NA's

```python
data = data.dropna()

# Check number of missing values
print("Count of na's:", data.isnull().values.sum())

# Check number of rows with missing values
print("Cases with missing values:", data.isna().any(axis=1).sum())
```

# Exploratory Data Analysis

### Explore variables with histograms


```python
data["age"].plot.hist(bins = 20)
```

### Explore the bivariate relationships with Joint Plot

```python
# import stats for pearson's r
import scipy.stats as stats

# selects the styling of the grid
# sns.set(style="white", color_codes=True)

xVar = "age"
yVar = "totlngth"

# the actual plot 
graph = sns.jointplot(data=data, 
                     x=xVar, 
                     y=yVar)

# finding pearson's r and p-val
r, p = stats.pearsonr(x=data[xVar],
                      y=data[yVar])

# legend properties
phantom, = graph.ax_joint.plot([], [], linestyle="", alpha=0)

# putting a legend on the ax joint grid
graph.ax_joint.legend([phantom],['r={:f}, p={:f}'.format(r,p)])
```

### Explore with KDE Joint Plot

```python
# import stats for pearson's r
import scipy.stats as stats

# selects the styling of the grid
# sns.set(style="white", color_codes=True)

xVar = "age"
yVar = "totlngth"

# the actual plot 
graph = sns.jointplot(data=data, 
                     x=xVar, 
                     y=yVar,
                     kind="kde")

# finding pearson's r and p-val
r, p = stats.pearsonr(x=data[xVar],
                      y=data[yVar])

# legend properties
phantom, = graph.ax_joint.plot([], [], linestyle="", alpha=0)

# putting a legend on the ax joint grid
graph.ax_joint.legend([phantom],['r={:f}, p={:f}'.format(r,p)])
```

### Explore all bivariate relationships with Pair Plot

```python
sns.pairplot(data, hue = "sex")
```
